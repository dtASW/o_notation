```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### The exercise

| Complexity function | O - Notation | 
| --- | --- | 
| 1000*n | O(n) | 
| 200(n log n) | O(n log n) | 
| 10*n^{2} | O(n^2) | 
| 2^{n} | O(2^n) | 

Functions

```python
def lineal(x):
    y = 1000 *x
    return y

def logarithmic(x):
    y = 200*x*np.log(x)
    return y

def cuadratica(x):
    y = 10 *(x**2)
    return y

def exponencial(x):
    y = 2**x
    return y
```

Plotting

```python
x = np.linspace (1, 15,100) 
plt.plot(x,lineal(x), label='Linear')
plt.plot(x,logarithmic(x), label='n log n')
plt.plot(x,cuadratica(x), label='Quadratic')
plt.plot(x,exponencial(x), label='Exponential')
plt.title('Orders of complexity')
plt.xlabel('n')
plt.ylabel('T(n)')
plt.legend()
plt.show()
```
