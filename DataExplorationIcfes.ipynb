## Data Exploration
### Top 10 Schools with the best scores

```python
df.groupby(['COLE_NOMBRE_ESTABLECIMIENTO'])['PUNT_GLOBAL'].mean()
```

```python
len((df['COLE_NOMBRE_ESTABLECIMIENTO'].unique()))
```

```python
puntajes = df.groupby(['COLE_NOMBRE_ESTABLECIMIENTO'])['PUNT_GLOBAL'].mean()
puntajes = puntajes.sort_values(ascending=False)
print(puntajes)
colegios = puntajes.index.tolist()
puntajes = puntajes.tolist()

top_10_colegios = []
top_10_puntajes = []

for i in range(10):
    top_10_colegios.append(colegios[i])
    top_10_puntajes.append(puntajes[i])

#Reversing list for better data visualization
top_10_colegios.reverse()
top_10_puntajes.reverse()


#Ploting
plt.barh(top_10_colegios, top_10_puntajes)
plt.title('Colegios con mejores puntajes del ICFES 2019')
plt.xlabel('Puntajes')
plt.show()
```

A qué regiones pertenecen esos colegios:

```python
for colegio in top_10_colegios:
    df_filtered = df.query(f'COLE_NOMBRE_ESTABLECIMIENTO == "{colegio}"')
    print(df_filtered['ESTU_DEPTO_RESIDE'].unique())
    print(df_filtered['COLE_NATURALEZA'].unique())
```

## Tiempo que pasan en internet

```python
df['ESTU_DEDICACIONINTERNET'].value_counts()
```

```python
internet_puntajes = df.groupby(['ESTU_DEDICACIONINTERNET'])['PUNT_GLOBAL'].mean()
internet_puntajes = internet_puntajes.sort_values(ascending=False)
print(internet_puntajes)
internet = internet_puntajes.index.to_list()
media_puntaje = internet_puntajes.to_list()

internet.reverse()
media_puntaje.reverse()

plt.barh(internet, media_puntaje)
plt.show()
```
